```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
X_class1 = np.random.randn(50, 2) + np.array([2, 2])
X_class2 = np.random.randn(50, 2) + np.array([-2, -2])

X = np.vstack((X_class1, X_class2))
y = np.array([1]*50 + [-1]*50)
```

```python
def perceptron(X, y, lr=0.1, epochs=1000):
    w1 = w2 = b = 1  # initial weights and bias

    for _ in range(epochs):
        for i in range(X.shape[0]):
            z = w1*X[i][0] + w2*X[i][1] + b
            if z*y[i] < 0:
                w1 += lr * y[i] * X[i][0]
                w2 += lr * y[i] * X[i][1]
                b += lr * y[i]
    return w1, w2, b
```

```python
w1, w2, b = perceptron(X, y)

print("Trained weights and bias:")
print("w1:", w1, "w2:", w2, "b:", b)
```

```python
plt.scatter(X_class1[:,0], X_class1[:,1], color='blue', label='Class 1')
plt.scatter(X_class2[:,0], X_class2[:,1], color='red', label='Class -1')
```

```python
plt.scatter(X_class1[:,0], X_class1[:,1], color='blue')
plt.scatter(X_class2[:,0], X_class2[:,1], color='red')

x_vals = np.linspace(X[:,0].min()-1, X[:,0].max()+1, 100)
y_vals = -(w1*x_vals + b)/w2
plt.plot(x_vals, y_vals, color='green')

plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.title("Perceptron Classification")
plt.show()
```
